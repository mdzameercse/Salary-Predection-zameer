```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('salary_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
df = data.copy()
```

```python
print("Data Sebelum Pemeriksaan Duplikat:")
print(df.shape)
```

```python
duplicated_rows = df[df.duplicated(keep=False)]
duplicated_rows
```

```python
df = df.drop_duplicates()
```

```python
print("\nData Setelah Pemeriksaan Duplikat:")
print(df.shape)
```

```python
df.isna().sum()
```

```python
X = df['experience_years']
X.head()
```

```python
y = df['salary']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.75,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train_reshape = X_train.values.reshape(-1, 1)
X_test_reshape = X_test.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_lr = LinearRegression()
model_lr.fit(X_train_reshape, y_train)
```

```python
y_test_pred = model_lr.predict(X_test_reshape)
y_test_pred
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_test_pred, color='red', label='Predicted')
plt.plot(X_test, y_test_pred, color='green', label='Regression Line', linewidth=2) 
plt.title('Linear Regression: Test Data vs Test Prediction')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

#i alredy have knowledege about this i am learn this in our college ML lab that why i am done this 
y_test_pred = model_lr.predict(X_test_reshape)
y_train_pred = model_lr.predict(X_train_reshape)

mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print(f"""
Mean Squared Error:
  Train: {mse_train:.2f}
  Test : {mse_test:.2f}
  Gap  : {abs(mse_train - mse_test):.2f}
R^2 Score:
  Train: {r2_train:.2f}
  Test : {r2_test:.2f}
      """)
```

```python
print('Intercept of the Linear Regression model:',model_lr.intercept_)
print('Coefficient of the line Linear Regression:',model_lr.coef_)
```

```python
from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(X_train_reshape, y_train)
```

```python
y_pred_dt = model_dt.predict(X_test_reshape)

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred_dt, color='red', label='Predicted (Decision Tree)')
plt.title('Decision Tree: Test Data vs Test Prediction')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_test_pred = model_dt.predict(X_test_reshape)
y_train_pred = model_dt.predict(X_train_reshape)

mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print(f"""
Mean Squared Error:
  Train: {mse_train:.2f}
  Test : {mse_test:.2f}
  Gap  : {abs(mse_train - mse_test):.2f}
R^2 Score:
  Train: {r2_train:.2f}
  Test : {r2_test:.2f}
      """)
```

```python
import joblib
joblib.dump(model_dt, 'decision_tree_model.pkl')
```

```python
loaded_model = joblib.load('decision_tree_model.pkl')


y_pred_load = loaded_model.predict(X_test_reshape)

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred_load, color='red', label='Predicted (Decision Tree)')
plt.legend()
plt.show()
```

### Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(X_train_reshape, y_train)
```

```python
y_pred_rf = model_rf.predict(X_test_reshape)

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred_rf, color='red', label='Predicted (Random Forest)')
plt.title('Random Forest: Test Data vs Test Prediction')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python

y_test_pred = model_rf.predict(X_test_reshape)
y_train_pred = model_rf.predict(X_train_reshape)

mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

print(f"""
Mean Squared Error:
  Train: {mse_train:.2f}
  Test : {mse_test:.2f}
  Gap  : {abs(mse_train - mse_test):.2f}
R^2 Score:
  Train: {r2_train:.2f}
  Test : {r2_test:.2f}
      """)
```

```python

joblib.dump(model_rf, 'random_forest_model.pkl')
```

```python

loaded_rf_model = joblib.load('random_forest_model.pkl')
y_pred_load = loaded_rf_model.predict(X_test_reshape)

plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.scatter(X_test, y_pred_load, color='red', label='Predicted (Random Forest)')
plt.title('Loaded Random Forest Model Predictions')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
